```python
from scipy.interpolate import interp1d
from numpy.polynomial.polynomial import Polynomial

return_speed = 20.1 # in mm/s
waggle_speed = 15.04

def bee_trajectory_generator(nb_waggle):
    #parameter
    waggle_angle = 0 # 0° nach oben, ccw
    waggle_divergence = 29.2667 * np.pi / 180 #default im paper: 32°
    waggle_divergence = 25 * np.pi / 180 #default im paper: 32°
    waggle_speed = 15.04 # in mm/s vortrieb des waggle runs
    waggle_duration = 43 * 2 * nb_waggle # in ms
    waggle_freq = 13 # in Hz
    waggle_amp_a = 6 * np.pi / 180 # in [0,2*pi] #hier peak to baseline angeben!!! 5,9
    return_speed = 20.1 # in mm/s
    return_duration = 2130 # in ms
    exc  = 15. #22
    waggle_ramp_length = 150 #[ms]

    resampling_steps = 700
    resampling_coeff = resampling_steps / return_duration #steps per millisecond
    
    #Definition of the orientation function, this three polynome give us the orientation and speed 
    orientation_polynomial_coef = np.array([0.8562, -3.572, 7.517, -4.581]) * resampling_coeff
    orientation_function = Polynomial(orientation_polynomial_coef) # return_orientation_function

    velocityX_polynomial_coef = np.array([0.03409, -0.1805, 0.9225, -1.466, 0.728]) * resampling_coeff
    velocityX_function = Polynomial(velocityX_polynomial_coef) # return_velocityX_function

    velocityY_polynomial_coef = np.array([-0.0337, 0.1027, -0.04566, -0.2951, 0.2683]) * resampling_coeff
    velocityY_function = Polynomial(velocityY_polynomial_coef) # return_velocityY_function
    
    # ?????
    t=np.linspace(0,1,return_duration)
    alpha = np.exp(10)/100
    return_orientation_weights = ((((t-0.5)**2)+alpha))
    return_orientation_weights = return_orientation_weights / ((np.sum(return_orientation_weights) / return_duration))

    # Return a scale for orientation
    t = np.linspace(0,1,return_duration)
    f = orientation_function(t)
    turn_scale = (360-(waggle_divergence/2)*180/np.pi)/sum(f)
    
    def getWaggleVelocity(t, left):
        v = np.array([0., 0., 0.])
        num_periods = ((waggle_duration/1000)*waggle_freq)
        sample_points_for_one_period = 1000 / waggle_freq

        if left:
            sign = 1
        else:
            sign = -1

        ramp = 1
        num_periods = (waggle_duration/1000.0) * waggle_freq
        sample_points_for_one_period = 1000.0 / waggle_freq

        if t < sample_points_for_one_period : 
            ramp = t / sample_points_for_one_period
        else:
            if t > (sample_points_for_one_period * (num_periods - 1)) :
                ramp = (waggle_duration - t) / sample_points_for_one_period

        t_prcnt = (t/waggle_duration)
        num_periods = ((waggle_duration/1000)*waggle_freq)

        v[0] = waggle_speed * np.cos(waggle_angle + sign*waggle_divergence/2) / 1000
        v[1] = waggle_speed * np.sin(waggle_angle + sign*waggle_divergence/2) / 1000
        v[2] = waggle_amp_a * ramp * np.cos(num_periods * t_prcnt * 2 * np.pi) / (1000 / (waggle_freq * 4) / (np.pi*0.5))

        return v
    
    def getReturnVelocity(t, Ocur, left):
        v = np.array([0., 0., 0.])

        if left:
            sign = 1
        else:
            sign = -1

        ramp = 1;
        d = 100;
        if t < d:
            ramp = t / d;

        unit_t = t / return_duration;
        v[2] = return_orientation_weights[int(min(return_duration, np.floor(t)+1))] * turn_scale*sign*ramp*np.pi*orientation_function(unit_t)/180

        v_fwd = turn_scale*ramp*velocityX_function(unit_t)
        v_swd = turn_scale*ramp*sign*velocityY_function(unit_t)

        c = np.cos(Ocur)
        s = np.sin(Ocur)

        v[0] = c*v_fwd - s*v_swd # + exc*(np.cos(Ocur+v[2])-np.cos(Ocur))
        v[1] = s*v_fwd + c*v_swd # + exc*(np.sin(Ocur+v[2])-np.sin(Ocur))
        return v
    
    current_position = np.array([0, 0, waggle_divergence/2])
    position = current_position

    for t in range(0, return_duration-1):
        q = getReturnVelocity(t, current_position[2], True)
        current_position = current_position + q
        position = np.vstack([position, current_position])

    current_position = np.array([0, 0, current_position[2]])
    position_waggle = np.array([0, 0, current_position[2]])

    for t in range(0, waggle_duration):
        q = getWaggleVelocity(t, False)
        current_position = current_position + q
        position_waggle = np.vstack([position_waggle, current_position])
        
    position_waggle[:,0] = position_waggle[:,0] - exc*np.cos(position_waggle[:,2]);
    position_waggle[:,1] = position_waggle[:,1] - exc*np.sin(position_waggle[:,2]);
    position_waggle[:,0] = position_waggle[:,0] - position_waggle[0,0]
    position_waggle[:,1] = position_waggle[:,1] - position_waggle[0,1]
    
    return position, position_waggle
    
    

def stretch_and_interpolate_return(stretch_function, goals, vector):
    """
    This function return a scipy interpolate object, you input a set of points of dim (K, 3) with K points
    who represents a trajectory (x,y,theta). this function will stretch your set of point to arrive to a desired point
    and interpolate all the values to return you an interpolation defime between 0 and 1 
    
    stretch_function : the function who will stretch yout points
    This function will be used only between 0 and 1 and will be normalized.
    
    (x_goal, y_goal, theta_goal) : goal point
    
    vector : your vector of points
    """
    new_return_speed = sum(np.linalg.norm(vector[:-1,:-1] - vector[1:,:-1], axis=1)) / (return_duration/1000) # mm.s-1
    
    def stretch_function_normalize(n_step, function):
        t = np.linspace(0, 1, n_step)
        farr = t

        for i in range(t.shape[0]):
            farr[i] = f(t[i])

        integral_f = sum(farr)

        farr = farr/integral_f

        return farr
    
    x_goal, y_goal, theta_goal = goals
    
    #Definition of all the deltas to compensate
    deltas = np.array([vector[-1, 0] - x_goal, vector[-1, 1] - y_goal, vector[-1, 2] - theta_goal])
    
    #normalization of the given stretch function (integrale[0->1] = 1)
    stretch_function_normalized = stretch_function_normalize(vector.shape[0], stretch_function)
    
    #Out vector
    vector_rescaled = np.array(vector)
    
    #Stretiching
    for i in range(vector.shape[0]):
        vector_rescaled[i, 0] = vector[i, 0] - sum(deltas[0] * stretch_function_normalized[:i])
        vector_rescaled[i, 1] = vector[i, 1] - sum(deltas[1] * stretch_function_normalized[:i])
        vector_rescaled[i, 2] = vector[i, 2] - sum(deltas[2] * stretch_function_normalized[:i])
      
    #Length of the trajectory for deducting the duration
    integral_length = np.zeros(vector_rescaled.shape[0]-1)
    for i in range(0, vector_rescaled.shape[0]-1):
        integral_length[i] = sum(np.linalg.norm(vector_rescaled[:i+1,:-1] - vector_rescaled[1:i+2,:-1], axis=1))
    integral_length[0] = 0.0001 
    
    total_path_length = sum(np.linalg.norm(vector_rescaled[:-1,:-1] - vector_rescaled[1:,:-1], axis=1))
    total_return_duration = 1
    print("1 : ", integral_length, " 2 = ", total_path_length)
    
    #We stretch the time vector to have a uniform repartition of the point (ex 1ms = 1mm)
    # Function : tn = (sum(dn)_0^K * Ttot)/Dtot
    # stretched_time_vector = np.zeros(stretch_function_normalized.shape[0])
    stretched_time_vector = np.zeros(stretch_function_normalized.shape[0])
    for i in range(1, stretched_time_vector.shape[0]):
        stretched_time_vector[i] = integral_length[i-1] * total_return_duration / integral_length[-1]
    vector_rescaled[-1,:] = [x_goal, y_goal, theta_goal]
    
    Xinterpolation = interp1d(stretched_time_vector, np.reshape(vector_rescaled[:,0], -1), kind='cubic')
    Yinterpolation = interp1d(stretched_time_vector, np.reshape(vector_rescaled[:,1], -1), kind='cubic')
    Thetainterpolation = interp1d(stretched_time_vector, np.reshape(vector_rescaled[:,2], -1), kind='cubic')
    
    return Xinterpolation, Yinterpolation, Thetainterpolation, total_return_duration

def generate_waggle_function(n_max, endpoint):
    _, position_waggle = bee_trajectory_generator(n_max)
    
    duration = position_waggle.shape[0] / 1000.
    start_point = position_waggle[0,:]
    end_point = position_waggle[-1,:]
    end_point[2] = np.pi * 2 - endpoint
    
    waggle_function_list = []
    
    np_for_ajust = 50
    time_for_ajust = np_for_ajust / 1000.
    
    position_waggle = position_waggle[:-np_for_ajust,:]
    position_waggle = np.vstack([position_waggle, end_point])
    
    time_vector = np.linspace(0, duration - time_for_ajust, position_waggle.shape[0]-1)
    time_vector = np.hstack([time_vector, duration])
    
    Xinterpolation = interp1d(time_vector, np.reshape(position_waggle[:,0], -1), kind='linear')
    Yinterpolation = interp1d(time_vector, np.reshape(position_waggle[:,1], -1), kind='linear')
    Thetainterpolation = interp1d(time_vector, np.reshape(position_waggle[:,2], -1), kind='linear')
    
    waggle_function_list = [[Xinterpolation, Yinterpolation, Thetainterpolation, duration]] + waggle_function_list
    
    for i in range(n_max-1, 0,-1):
        _, position_waggle = bee_trajectory_generator(i)
        time_vector = np.linspace(0, position_waggle.shape[0]/1000, position_waggle.shape[0])
        position_waggle = np.vstack([position_waggle, end_point])
        time_vector = np.hstack([time_vector, duration])
        
        Xinterpolation = interp1d(time_vector, np.reshape(position_waggle[:,0], -1), kind='linear')
        Yinterpolation = interp1d(time_vector, np.reshape(position_waggle[:,1], -1), kind='linear')
        Thetainterpolation = interp1d(time_vector, np.reshape(position_waggle[:,2], -1), kind='linear')
        
        waggle_function_list = [[Xinterpolation, Yinterpolation, Thetainterpolation, duration]] + waggle_function_list
    
    return waggle_function_list
        
exc =15.
waggle_count = 15

svg_size = np.array([60,60])

position, position_waggle = bee_trajectory_generator(waggle_count)

plt.plot(position[:,0], position[:,1], 'r-')
plt.title("Return trajectory generated by Tim")
plt.show()
plt.scatter(position_waggle[:,0], position_waggle[:,1], c=position_waggle[:,2])
plt.title("Waggle trajectory generated by Tim")
plt.show()

#Stretch function
f = lambda x : x if 0 <= x <= 0.5 else ((1-x) if 0.5 < x <= 1 else 0)

#Stretch the function to fit the waggle
x_wanted = -position_waggle[-1,0]
y_wanted = 0.
Xinter, Yinter, Thetainter, duration = stretch_and_interpolate_return(f, (x_wanted, y_wanted, position[-1, 2]), position)

position_waggle[:,0] = position_waggle[:,0] + position[-1,0]
position_waggle[:,1] = position_waggle[:,1] + position[-1,1]

t = np.linspace(0, duration,1000)

waggle_return = dict()

plt.scatter(Xinter(t), Yinter(t), c=Thetainter(t))
plt.show()

waggle_return['inter_x_bee'] = Xinter
waggle_return['inter_y_bee'] = Yinter
waggle_return['inter_theta_bee'] = Thetainter
waggle_return['duration'] = duration

np.save('return_origin.npy', waggle_return)


waggle_list = generate_waggle_function(waggle_count, Thetainter(0))

t = np.linspace(0, duration,1000)
trajectory_return = np.transpose(np.array([Xinter(t), Yinter(t), Thetainter(t)]))

t = np.linspace(0, waggle_list[14][3],1000)
trajectory_waggle = np.transpose(np.array([waggle_list[14][0](t) + x_wanted,
                                           waggle_list[14][1](t),
                                           waggle_list[14][2](t)]))


robot_trajectory_return = np.transpose(np.array([trajectory_return[:,0] + exc*np.cos(trajectory_return[:,2]), 
                                                 trajectory_return[:,1] + exc*np.sin(trajectory_return[:,2]),
                                                 trajectory_return[:,2]]))
robot_trajectory_waggle = np.transpose(np.array([trajectory_waggle[:,0] + exc*np.cos(trajectory_waggle[:,2]), 
                                                 trajectory_waggle[:,1] + exc*np.sin(trajectory_waggle[:,2]),
                                                 trajectory_waggle[:,2]]))

t = np.linspace(0, duration,1000)
trajectory_return_bis = np.transpose(np.array([Xinter(t) + trajectory_waggle[-1,0],
                                               -Yinter(t) + trajectory_waggle[-1,1], 
                                               -Thetainter(t)]))
t = np.linspace(0, waggle_list[14][3],1000)
trajectory_waggle_bis = np.transpose(np.array([waggle_list[14][0](t) + trajectory_return_bis[-1,0],
                                               -waggle_list[14][1](t) + trajectory_return_bis[-1,1],
                                               -waggle_list[14][2](t)]))

robot_trajectory_return_bis = np.transpose(np.array([trajectory_return_bis[:,0] + exc*np.cos(trajectory_return_bis[:,2]), 
                                                     trajectory_return_bis[:,1] + exc*np.sin(trajectory_return_bis[:,2]),
                                                     trajectory_return_bis[:,2]]))
robot_trajectory_waggle_bis = np.transpose(np.array([trajectory_waggle_bis[:,0] + exc*np.cos(trajectory_waggle_bis[:,2]), 
                                                     trajectory_waggle_bis[:,1] + exc*np.sin(trajectory_waggle_bis[:,2]),
                                                     trajectory_waggle_bis[:,2]]))


#Find the centerpoint
min_x_robot_return = np.amin(robot_trajectory_return[:,0])
min_y_robot_return = np.amin(robot_trajectory_return[:,1])
max_x_robot_return = np.amax(robot_trajectory_return[:,0])
max_y_robot_return = np.amax(robot_trajectory_return[:,1])

min_x_robot_return_bis = np.amin(robot_trajectory_return_bis[:,0])
min_y_robot_return_bis = np.amin(robot_trajectory_return_bis[:,1])
max_x_robot_return_bis = np.amax(robot_trajectory_return_bis[:,0])
max_y_robot_return_bis = np.amax(robot_trajectory_return_bis[:,1])

min_x_robot_waggle = np.amin(robot_trajectory_waggle[:,0])
min_y_robot_waggle = np.amin(robot_trajectory_waggle[:,1])
max_x_robot_waggle = np.amax(robot_trajectory_waggle[:,0])
max_y_robot_waggle = np.amax(robot_trajectory_waggle[:,1])

min_x_robot_waggle_bis = np.amin(robot_trajectory_waggle_bis[:,0])
min_y_robot_waggle_bis = np.amin(robot_trajectory_waggle_bis[:,1])
max_x_robot_waggle_bis = np.amax(robot_trajectory_waggle_bis[:,0])
max_y_robot_waggle_bis = np.amax(robot_trajectory_waggle_bis[:,1])

mini_x = min(min_x_robot_return, min_x_robot_return_bis, min_x_robot_waggle, min_x_robot_waggle_bis)
maxi_x = max(max_x_robot_return, max_x_robot_return_bis, max_x_robot_waggle, max_x_robot_waggle_bis)

mini_y = min(min_y_robot_return, min_y_robot_return_bis, min_y_robot_waggle, min_y_robot_waggle_bis)
maxi_y = max(max_y_robot_return, max_y_robot_return_bis, max_y_robot_waggle, max_y_robot_waggle_bis)

center = np.array([-(mini_x + maxi_x)/2 + svg_size[0]/2, -(mini_y + maxi_y)/2 + svg_size[1]/2])

trajectory_return[:, :2] = trajectory_return[:, :2] + center 
trajectory_waggle[:, :2] = trajectory_waggle[:, :2] + center
trajectory_return_bis[:, :2] = trajectory_return_bis[:, :2] + center
trajectory_waggle_bis[:, :2] = trajectory_waggle_bis[:, :2] + center

robot_trajectory_return[:, :2] = robot_trajectory_return[:, :2] + center 
robot_trajectory_waggle[:, :2] = robot_trajectory_waggle[:, :2] + center
robot_trajectory_return_bis[:, :2] = robot_trajectory_return_bis[:, :2] + center
robot_trajectory_waggle_bis[:, :2] = robot_trajectory_waggle_bis[:, :2] + center

plt.figure(figsize=(20,10))
plt.scatter(trajectory_return[:,0], trajectory_return[:,1], c=trajectory_return[:,2], cmap='magma')
plt.scatter(trajectory_waggle[:,0], trajectory_waggle[:,1], c=trajectory_waggle[:,2], cmap='magma')
plt.scatter(trajectory_return_bis[:,0], trajectory_return_bis[:,1], c=trajectory_return_bis[:,2], cmap='magma')
plt.scatter(trajectory_waggle_bis[:,0], trajectory_waggle_bis[:,1], c=trajectory_waggle_bis[:,2], cmap='magma')

plt.scatter(robot_trajectory_return[:,0], robot_trajectory_return[:,1], c=robot_trajectory_return[:,2])
plt.scatter(robot_trajectory_waggle[:,0], robot_trajectory_waggle[:,1], c=robot_trajectory_waggle[:,2])
plt.scatter(robot_trajectory_return_bis[:,0], robot_trajectory_return_bis[:,1], c=robot_trajectory_return_bis[:,2])
plt.scatter(robot_trajectory_waggle_bis[:,0], robot_trajectory_waggle_bis[:,1], c=robot_trajectory_waggle_bis[:,2])

plt.show()

plt.plot(trajectory_return[:,2])
plt.plot(trajectory_waggle[:,2])
plt.plot(trajectory_return_bis[:,2])
plt.plot(trajectory_waggle_bis[:,2])
plt.show()

```

```python
import json
from scipy.interpolate import interp1d
import numpy as np

def return_generator(origin_trajectory, speed_ramp, stretch_function, goals, config_json):
    
    with open(config_json, 'r') as myfile:
        data=myfile.read()
    # parse file
    config = json.loads(data)
    return_speed = config['return_speed_mms-1']
    waggle_divergence = config['waggle_divergence_rad']
    exc = config['exc_mm']
    steps = 2000
    
    return_interpolation = np.load(origin_trajectory, allow_pickle=True).item()
        
    def stretch_function_normalize(n_step, function):
        t = np.linspace(0, 1, n_step)
        farr = t

        for i in range(t.shape[0]):
            farr[i] = f(t[i])

        integral_f = sum(farr)

        farr = farr/integral_f

        return farr
    
    x_goal, y_goal, theta_goal = goals
    
    t = np.linspace(0, 1, steps)
    vector = np.transpose(np.array([return_interpolation['inter_x_bee'](t), 
                                    return_interpolation['inter_y_bee'](t),
                                    return_interpolation['inter_theta_bee'](t)]))
    
    delta_theta_init = waggle_divergence / 2 - vector[0, 2]
    vector[:, 2] = vector[:, 2] + delta_theta_init
    
    #Definition of all the deltas to compensate
    deltas = np.array([vector[-1, 0] - x_goal, vector[-1, 1] - y_goal, vector[-1, 2] - theta_goal])
    
    #normalization of the given stretch function (integrale[0->1] = 1)
    stretch_function_normalized = stretch_function_normalize(vector.shape[0], stretch_function)
    
    #Out vector
    vector_rescaled = np.array(vector)
    
    #Stretiching
    for i in range(vector.shape[0]):
        vector_rescaled[i, 0] = vector[i, 0] - sum(deltas[0] * stretch_function_normalized[:i])
        vector_rescaled[i, 1] = vector[i, 1] - sum(deltas[1] * stretch_function_normalized[:i])
        vector_rescaled[i, 2] = vector[i, 2] - sum(deltas[2] * stretch_function_normalized[:i])
    
    #Length of the trajectory for deducting the duration
    integral_length = np.zeros(vector_rescaled.shape[0]-1)
    for i in range(0, vector_rescaled.shape[0]-1):
        integral_length[i] = sum(np.linalg.norm(vector_rescaled[:i+1,:-1] - vector_rescaled[1:i+2,:-1], axis=1))
    integral_length[0] = 0.0001 
    
    total_path_length = sum(np.linalg.norm(vector_rescaled[:-1,:-1] - vector_rescaled[1:,:-1], axis=1))
    total_return_duration = 1
    
    #We stretch the time vector to have a uniform repartition of the point (ex 1ms = 1mm)
    # Function : tn = (sum(dn)_0^K * Ttot)/Dtot
    # stretched_time_vector = np.zeros(stretch_function_normalized.shape[0])
    stretched_time_vector = np.zeros(stretch_function_normalized.shape[0])
    for i in range(1, stretched_time_vector.shape[0]):
        stretched_time_vector[i] = integral_length[i-1] * total_return_duration / integral_length[-1]
    vector_rescaled[-1,:] = [x_goal, y_goal, theta_goal]
    
    speed_ramp_tab = np.zeros(steps)
    for i in range(steps):
        speed_ramp_tab[i] = speed_ramp(t[i]) * return_speed
        
    duration = integral_length[-1] / np.trapz(speed_ramp_tab, dx=1/steps)
    
    stretched_time_vector = stretched_time_vector * duration
    
    Xinterpolation = interp1d(stretched_time_vector, np.reshape(vector_rescaled[:,0], -1), kind='cubic')
    Yinterpolation = interp1d(stretched_time_vector, np.reshape(vector_rescaled[:,1], -1), kind='cubic')
    Thetainterpolation = interp1d(stretched_time_vector, np.reshape(vector_rescaled[:,2], -1), kind='cubic')
    
    interpolation_dict = {'inter_x_bee': Xinterpolation,
                          'inter_y_bee': Yinterpolation,
                          'inter_theta_bee': Thetainterpolation,
                          'duration': duration}
    
    return interpolation_dict
  
    
#Map any angle between -pi and pi
angle_mapper = lambda x : ( x - np.pi ) % ( 2 * np.pi ) - np.pi

path_trajectory = 'return_origin.npy'
waggle_path = 'waggle_1-14.npy'
config_json = 'config.json'

ramp = 0.1
f = lambda x : (x if 0 <= x <= ramp else ((1-x) if 1-ramp < x <= 1 else ((ramp) if ramp < x <= 1-ramp else 0)))*1/ramp

#Stretch function
f_stretch = lambda x : x if 0 <= x <= 0.5 else ((1-x) if 0.5 < x <= 1 else 0)

waggle_function_list = np.load(waggle_path, allow_pickle=True)
x_max = -waggle_function_list[0]['inter_x_bee'](waggle_function_list[0]['duration'])
theta_goal = 2*np.pi + waggle_function_list[0]['inter_theta_bee'](0)
y_goal = 0.0
goal = (x_max, y_goal, theta_goal)
print(goal)

return_function = return_generator(path_trajectory, f, f_stretch, goal, config_json)
```

```python
import copy
import svgwrite

waggle_function = waggle_function_list[14]

t_return = np.linspace(0, return_function['duration'], int(return_function['duration']*1000))
t_waggle = np.linspace(0, waggle_function['duration'], int(waggle_function['duration']*1000))

waggle = np.transpose(np.array([waggle_function['inter_x_bee'](t_waggle), 
                                waggle_function['inter_y_bee'](t_waggle),
                                waggle_function['inter_theta_bee'](t_waggle)]))

waggle_bis = copy.copy(waggle)
waggle_bis[:,1:] = -waggle_bis[:,1:]

return_loop = np.transpose(np.array([return_function['inter_x_bee'](t_return[2:]), 
                                     return_function['inter_y_bee'](t_return[2:]),
                                     return_function['inter_theta_bee'](t_return[2:])]))

return_loop_bis = copy.copy(return_loop)
return_loop_bis[:,1:] = -return_loop_bis[:,1:]

def trajectory_merger(trajectory_tab):
    
    nb_trajectory = len(trajectory_tab)
    
    merged_trajectory = trajectory_tab[0]
        
    for i in range(1, nb_trajectory):
        last_position = merged_trajectory[-1]
        offset_trajectory = trajectory_tab[i]
        offset_trajectory[:,:2] = offset_trajectory[:,:2] + last_position[:2] 
        merged_trajectory = np.vstack([merged_trajectory, offset_trajectory])
     
    return merged_trajectory

def return_robot_trajetory(trajectory):
    exc = 15.
    robot_traj = copy.copy(trajectory)
    robot_traj[:,0] = trajectory[:,0] + exc * np.cos(trajectory[:,2])
    robot_traj[:,1] = trajectory[:,1] + exc * np.sin(trajectory[:,2])
    return robot_traj

def center_trajectory(trajectory, averx=None, avery=None):
    minx = np.amin(trajectory[:,0])
    maxx = np.amax(trajectory[:,0])
    miny = np.amin(trajectory[:,1])
    maxy = np.amax(trajectory[:,1])
    
    if averx is None or avery is None:
        averx = (maxx + minx)/2
        avery = (maxy + miny)/2
    
    trajectory[:,0] = trajectory[:,0] - averx
    trajectory[:,1] = trajectory[:,1] - avery
    
    return trajectory, averx, avery

def export_to_svg(trajectory):
    
    def coord_to_string(command, x, y):
        return command + ' ' + str(x)+','+str(y)
    
    dwg = svgwrite.Drawing('dancepath.svg', size=('60mm', '60mm'), viewBox=('-30 -30 30 30'))
    dwg.add(dwg.line(start=(-30, -30), 
                     end=(30, -30), 
                     stroke=svgwrite.rgb(10, 10, 10, '%')))
    dwg.add(dwg.line(start=(-30, -30), 
                     end=(-30, 30), 
                     stroke=svgwrite.rgb(10, 10, 10, '%')))
    
    path = dwg.path(d=coord_to_string('M', trajectory[0,0], trajectory[0,1]))
    
    steps = 10
    for i in range(steps, trajectory.shape[0], steps):
        path.push(coord_to_string('L', trajectory[i,0], trajectory[i,1]))
    #    dwg.add(dwg.line(start=(trajectory[prev_i,0], trajectory[prev_i,1]), 
    #                 end=(trajectory[i,0], trajectory[i,1]), 
    #                 stroke=svgwrite.rgb(10, 10, 10, '%')))
        
    path.push(coord_to_string('L', trajectory[0,0], trajectory[0,1]))
    dwg.add(path)
    dwg.save()
    
tab = [return_loop, waggle, return_loop_bis, waggle_bis]

traj = trajectory_merger(tab)

#Map any angle between -pi and pi
angle_mapper = lambda x : ( x - np.pi ) % ( 2 * np.pi ) - np.pi
angle_mapper_vectorize = np.vectorize(angle_mapper)

traj[:,2] = angle_mapper_vectorize(traj[:,2])
plt.figure(figsize=(20,20))
plt.plot(traj[:,0], traj[:,1])
plt.show()

plt.plot(traj[:,2])
plt.show()

plt.figure(figsize=(20,20))
robot = return_robot_trajetory(traj)
robot, averx, avery = center_trajectory(robot)
traj, _, _ = center_trajectory(traj, averx, avery)
plt.scatter(traj[:,0], traj[:,1], c=traj[:,2], cmap='magma')
plt.scatter(robot[:,0], robot[:,1], c=robot[:,2])
plt.show()

export_to_svg(robot)

```
